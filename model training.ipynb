```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_cleaned.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
##drop month,day and yyear
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
# Independent and Deoendent feature
x = df.drop('FWI',axis=1)
y = df['FWI']
```

```python
x.head()
```

```python
y
```

```python
# Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
## Feature Selection based on correlaltion
x_train.corr()
```

```python
# Check for multicolinearity
plt.figure(figsize=(10,8))
corr = x_train.corr()
sns.heatmap(corr , annot=True)
```

```python
x_train.corr()
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(x_train,0.85)
```

corr_features

```python
## drop features when correlation is more than 0.85 
x_train.drop(corr_features,axis=1,inplace=True)
x_test.drop(corr_features,axis=1,inplace=True)
x_train.shape,x_test.shape
```

# Feature scelling and standerdization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_sclaed = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_sclaed
```

# Box plot to understand the Effect of Standerd scaler

```python
plt.subplots(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=x_train)
plt.title('x_train before scalling')
plt.subplot(1, 2, 2)
sns.boxplot(data=x_train_sclaed)
plt.title('x_train after scalling')

```

# LInear regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
linreg.fit(x_train_sclaed,y_train)
y_pred=linreg.predict(x_test_scaled)
mean_absolute_error(y_test,y_pred)
r2_score  = r2_score(y_test,y_pred)
print('mean_absolute_error:',mean_absolute_error)
print('r2_score:',r2_score)
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(x_train_sclaed,y_train)
y_pred=lasso.predict(x_test_scaled) 
mean_absolute_error(y_test,y_pred)
r2_score  = r2_score(y_test,y_pred)
print('mean_absolute_error:',mean_absolute_error)
print('r2_score:',r2_score)
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

# Cross validation LAsso

```python
from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_absolute_error, r2_score

lassocv=LassoCV(cv=5)
lassocv.fit(x_train_sclaed,y_train)
```

```python
y_pred=lassocv.predict(x_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

# Ridge resression moldel

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(x_train_sclaed,y_train)
y_pred=ridge.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(x_train_sclaed,y_train)
y_pred=ridgecv.predict(x_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

```python
print("best score ",ridgecv.best_score_)
ridgecv.get_params()
```
